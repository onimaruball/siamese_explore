```python
def copy_files_to_new_folder(original_folder, dest_root):
    # สร้าง destination folder หากยังไม่มี
    os.makedirs(dest_root, exist_ok=True)

    file_count = 0  # ตัวแปรสำหรับนับจำนวนไฟล์ที่คัดลอก

    # คัดลอกไฟล์จาก original_folder ไปยัง dest_root
    for file_name in os.listdir(original_folder):
        file_path = os.path.join(original_folder, file_name)
        if os.path.isfile(file_path):
            dest_file_path = os.path.join(dest_root, file_name)
            shutil.copy(file_path, dest_file_path)  # คัดลอกไฟล์
            file_count += 1  # เพิ่มจำนวนไฟล์ที่คัดลอก

    # แสดงผลจำนวนไฟล์ที่คัดลอก
    print(f"Copied {file_count} files to {dest_root}")
            
import os

# ฟังก์ชันสำหรับเปลี่ยนชื่อโฟลเดอร์และไฟล์ภายในโฟลเดอร์
def rename_folder_and_files(original_folder, old_name, new_name):
    # ตรวจสอบว่า old_name อยู่ใน path หรือไม่
    if old_name in original_folder:
        # เปลี่ยนชื่อโฟลเดอร์
        new_folder_path = original_folder.replace(old_name, new_name)
        
        # เปลี่ยนชื่อโฟลเดอร์
        os.rename(original_folder, new_folder_path)
        print(f"Folder renamed from {original_folder} to {new_folder_path}")
        
        # เปลี่ยนชื่อไฟล์ภายในโฟลเดอร์
        for file_name in os.listdir(new_folder_path):
            new_file_name = file_name.replace(old_name, new_name)
            old_file_path = os.path.join(new_folder_path, file_name)
            new_file_path = os.path.join(new_folder_path, new_file_name)
            
            # เปลี่ยนชื่อไฟล์
            os.rename(old_file_path, new_file_path)
            print(f"Renamed {old_file_path} to {new_file_path}")
    else:
        print(f"{old_name} not found in {original_folder}")
```

```python
import os

# กำหนด path หลัก
dataset_path = "/home/ball/siamese_explore/dataset/"

# สร้างตัวแปรเพื่อเก็บจำนวนโฟลเดอร์ที่ขึ้นต้นด้วยแต่ละชื่อ
w_ls1_count = 0
w_ls2_count = 0
s_ss1_count = 0
s_ss2_count = 0

# ลูปผ่านทุกโฟลเดอร์ใน dataset_path
for root, dirs, files in os.walk(dataset_path):
    for dir_name in dirs:
        if dir_name.startswith("w_ls1"):
            w_ls1_count += 1
        elif dir_name.startswith("w_ls2"):
            w_ls2_count += 1
        elif dir_name.startswith("s_ss1"):
            s_ss1_count += 1
        elif dir_name.startswith("s_ss2"):
            s_ss2_count += 1

# แสดงผลลัพธ์
print(f"Number of folders starting with 'w_ls1': {w_ls1_count}")
print(f"Number of folders starting with 'w_ls2': {w_ls2_count}")
print(f"Number of folders starting with 's_ss1': {s_ss1_count}")
print(f"Number of folders starting with 's_ss2': {s_ss2_count}")
```
